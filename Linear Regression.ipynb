```python

X = X[["Cement"]]
y = y["Concrete compressive strength"]

# Model
model = LinearRegression()
model.fit(X, y)

# Prediction line
y_pred = model.predict(X)

# Plot
plt.scatter(X, y)
plt.plot(X, y_pred, color='red')

plt.xlabel("Cement")
plt.ylabel("Strength")
plt.title("Best Fit Line")
plt.show()
```
