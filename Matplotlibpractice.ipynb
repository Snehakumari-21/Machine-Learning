```python
import matplotlib.pyplot as plt
```

```python
x=[10,20,30,40]
y=[2,46,5,67]
plt.plot(x,y)
plt.show()
```

```python
plt.figure(figsize=(4,3))
plt.plot(x,y,color='blue', marker='o' , linestyle='--', linewidth=7 ,markersize=10)

plt.title("Title ")
plt.xlabel("x-  axis")
plt.ylabel("y-  axis")
plt.show()
```

```python
# MULTI LINE
x=[1,2,3,4,5]
y1=[10,20,25,35,45]
y2=[20,30,35,45,55]
plt.plot(x ,y1 , label= 'sales 2024')
plt.plot(x ,y2 , label= 'sales 2025')

plt.title("yearly sales")
plt.xlabel(" Months")
plt.ylabel(" sales")

plt.legend()
plt.show
```

```python
# barchart
x=['A','B','c','D']
y=[80,20,30,40]
plt.bar(x,y,width=0.3)
plt.title(" Bar chart")
plt.xlabel('grade')
plt.ylabel('mark')
plt.show()
```

```python
# matplotlib with panda - real data
# create df

import pandas as pd

# data
data={
    'Months':["jan","feb","mar","apr","may","jun"],
    'sales':[1200,346,5677,68,890,53]
}

df=pd.DataFrame(data)
df
```

```python
plt.bar(df["Months"],df["sales"])
plt.title("Matplotlib with Pandas")
plt.xlabel('month')
plt.ylabel('sales')

plt.savefig("Matplotlib_Sales_From_df.png")

plt.show()
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style="dark")

# Simulate data from a bivariate Gaussian
n = 10000
mean = [0, 0]
cov = [(2, .4), (.4, .2)]
rng = np.random.RandomState(0)
x, y = rng.multivariate_normal(mean, cov, n).T

# Draw a combo histogram and scatterplot with density contours
f, ax = plt.subplots(figsize=(6, 6))
sns.scatterplot(x=x, y=y, s=5, color=".15")
sns.histplot(x=x, y=y, bins=50, pthresh=.1, cmap="mako")
sns.kdeplot(x=x, y=y, levels=5, color="w", linewidths=1)
```
